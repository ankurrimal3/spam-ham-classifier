```python
import spacy
nlp=spacy.load('en_core_web_sm')
#importing spacy for removing stopword
```

```python
from nltk.stem.snowball import SnowballStemmer
s_stemmer = SnowballStemmer(language='english')
#import snowball stemmer for stemming
```

```python
corpus=[]
```

```python
#import regular expression
import re
for i in range(0, len(df)):
    #removing stop word
    review = re.sub('[^a-zA-Z]', ' ', df['message'][i])
    #converting into lowercase
    review = review.lower()
    #splitting individual message to word
    review = review.split()
    
    #select the ones for stemming which are not in stopword
    review = [s_stemmer.stem(word) for word in review if not word in nlp.Defaults.stop_words] 
    #joing the whole word to a sentence after cleaning
    review = ' '.join(review)
    #add each review to a corpus
    corpus.append(review)
    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
corpus
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
y=pd.get_dummies(df['label'])
y=y.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 101)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_pred,y_test)
```
